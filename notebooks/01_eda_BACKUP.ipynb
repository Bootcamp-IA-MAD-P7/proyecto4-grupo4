# EDA — Predicción de valoración de unicornios

Análisis exploratorio para el proyecto de regresión (Nivel Esencial).

```python
import sys
from pathlib import Path

ROOT = Path.cwd().parent
sys.path.insert(0, str(ROOT))

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from src.data.load import build_features, load_raw_dataset, prepare_modeling_frame

sns.set_theme(style="whitegrid")
```

```python
raw = load_raw_dataset()
df = build_features(raw)
X, y = prepare_modeling_frame(df)
model_df = X.copy()
model_df["valuation_usd"] = y
model_df.head()
```

```python
model_df.describe(include="all").T
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(model_df["valuation_usd"] / 1e9, bins=30, kde=True)
plt.xlabel("Valuation (B USD)")
plt.title("Distribución de la variable objetivo");
```

```python
numeric_cols = ["year_founded", "funding_usd", "company_age", "valuation_usd"]
plt.figure(figsize=(8, 6))
sns.heatmap(model_df[numeric_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de correlación");
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.scatterplot(data=model_df, x="funding_usd", y="valuation_usd", ax=axes[0], alpha=0.6)
axes[0].set_title("Funding vs Valuation")
sns.scatterplot(data=model_df, x="company_age", y="valuation_usd", ax=axes[1], alpha=0.6)
axes[1].set_title("Company age vs Valuation");
```

```python
top_industries = model_df["industry"].value_counts().head(8).index
subset = model_df[model_df["industry"].isin(top_industries)]
plt.figure(figsize=(10, 5))
sns.boxplot(data=subset, x="industry", y="valuation_usd")
plt.xticks(rotation=30, ha="right")
plt.title("Valoración por industria (top 8)");
```
